```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def sistema_1a_ordem(y, t, tau, u):
    """Modelo de um sistema de 1ª ordem: tau * dy/dt + y = u(t)
    Args:
        y (float): Estado atual (ex.: tensão, valor financeiro)
        t (float): Tempo
        tau (float): Constante de tempo (ex.: RC, tempo de decaimento)
        u (float): Entrada (ex.: degrau unitário)
    Returns:
        float: Derivada dy/dt
    """
    return (-y + u) / tau
```

```python
def resposta_analitica(t, tau, u, y0):
    """Solução analítica: y(t) = y0 * e^(-t/tau) + u * (1 - e^(-t/tau))
    Args:
        t (array): Vetor de tempos
        tau (float): Constante de tempo
        u (float): Entrada (constante)
        y0 (float): Condição inicial
    Returns:
        array: Resposta y(t)
    """
    return y0 * np.exp(-t / tau) + u * (1 - np.exp(-t / tau))
```

```python
# Exemplo fixo (baseado nos slides do Cap1)
print("Capítulo 1: Introdução a Sistemas Dinâmicos")
print("Exemplo: Sistema de 1ª ordem (ex.: circuito RC, decay financeiro)")
tau = 2.0    # Constante de tempo (segundos)
u = 1.0      # Entrada degrau (ex.: 1V ou R$1)
y0 = 0.0     # Condição inicial (ex.: capacitor descarregado)
t = np.linspace(0, 10, 100)  # Tempo de 0 a 10s
```

```python
# Solução numérica (usando scipy)
y_num = odeint(sistema_1a_ordem, y0, t, args=(tau, u)).flatten()
# Solução analítica
y_ana = resposta_analitica(t, tau, u, y0)
```

```python
# Gráfico do exemplo fixo
plt.plot(t, y_num, 'b-', label="Resposta Numérica (SciPy)")
plt.plot(t, y_ana, 'r--', label="Resposta Analítica")
plt.xlabel("Tempo (s)")
plt.ylabel("Saída y(t)")
plt.title("Resposta de um Sistema de 1ª Ordem (τ = 2s, u = 1)")
plt.grid(True)
plt.legend()
plt.show()

```
